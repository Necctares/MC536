## Conjunto de dados 1 - Media salarial por PUMAs (Área de Microdados de Uso Público) nos EUA (JSON)

```python
url = "https://ruby.datausa.io/api/data?CIP2=51&measure=ygcpop RCA,Total Population,Total Population MOE Appx,Average Wage,Average Wage Appx MOE,Record Count&drilldowns=PUMA&Record Count>=5&Workforce Status=true"
data = rq.get(url)
```

```python
#Tratando os dados obtidos
df = data.json()
df = df["data"]
df = pd.DataFrame(df)

#Filtrando os dados que nós queremos
df = df.filter(items=["ID PUMA","PUMA", "Year", "Total Population", "Average Wage", "ygcpop RCA"])
df
```
